```python
from fastai.vision.all import *
from ipywidgets import widgets
import pathlib
import matplotlib.pyplot as plt
import seaborn as sns
import IPython.display as ipd
```

### Pnevmonia kassaligni tashxis qilish

```python
pathlib.PosixPath = pathlib.WindowsPath
```

```python
pneumonia_model = load_learner('model/pnevmaniya_model.pkl')
```

```python
upload = widgets.FileUpload()
upload
```

```python
img_path = upload.value[-1].content.tobytes()
img = PILImage.create(img_path)
pred, pred_id, probs = pneumonia_model.predict(img)
print(f"It is a {pred}, accuracy: {probs[pred_id] * 100:.1f}%")
img.resize((350,350))
```
